# **Women Cloth Review Prediction with Multinomial Naive Bayes**

The Multinomial Naive Bayes classifier is suitable for classification with discrete features (e.g., word counts for text classification ). The Multinomial distribution normally requires integer feature counts. However, in practice, fractional counts such as tf-idf may also work.

# **Import Library**

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

# **Import Dataset**

```python
df = pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Teaching-Data/main/Women%20Clothing%20E-Commerce%20Review.csv')
```

# **Data PreProcesssing**

# **Get First Five Row of Dataset**

```python
df.head()
```

# **Get Information of Dataset**

```python
df.info()
```

# **Get Summary Statistics**

```python
df.describe()
```

# **Get shape of the Dataset**

```python
df.shape
```

```python
df.isna().sum()
```

```python
df[df['Review']==""]=np.NaN
```

```python
df['Review'].fillna("No Review",inplace=True)
```

```python
df.isna().sum()
```

```python
df['Review']
```

# **Define Target (y) and Feature (X)**

```python
df.columns
```

```python
X = df['Review']
```

```python
y = df['Rating']
```

```python
df['Rating'].value_counts()
```

# **Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,stratify=y,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# **Get Feature Text Conversion to Tokens**

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(lowercase = True, analyzer='word', ngram_range=(2,3), stop_words = 'english', max_features=5000)
```

```python
 X_train = cv.fit_transform(X_train)
```

```python
cv.get_feature_names_out()
```

```python
X_train.toarray()
```

```python
X_test = cv.fit_transform(X_test)
```

```python
cv.get_feature_names_out()
```

```python
X_test.toarray()
```

# **Get Model Train**

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train, y_train)
```

# **Get Model Prediction**

```python
y_pred = model.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

# **Get Probability of Each Predicted Class**

```python
model.predict_proba(X_test)
```

# **Get Model Evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
